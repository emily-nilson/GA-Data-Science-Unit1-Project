### Challenge 1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# start with two numbers at 1000, subtract 1 each time (in a loop) and multiply together
# then test if palindrome

num1 = 1000
num2 = 1000

for num in range (0,1000):
    potential_palindrome = num1 * num2
    reversed_potential_palindrome = ''.join(reversed(str(potential_palindrome)))
    num1 -= 1
    num2 -= 1 
    if str(potential_palindrome) == reversed_potential_palindrome:
        print(potential_palindrome, 'This is a palindrome!')
    else:
        print(potential_palindrome, False)
    
```

If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z).
```python
# use index [i] and [i+1] to move through the string to check for repeated chars
# start with empty output string to add results to
# use count to determine if chars repeated

def string_compressor(string):           # function with string input
    compressed = ''
    count = 1
    
    compressed += string[0]              # add first char to string, otherwise code misses it
    
    for i in range(0,len(string)-1):     # for each character in the string, i = index  
        if string[i] == string[i+1]:     # if two chars next to eachother same, add 1 to count
            count +=1
        else:
            if (count >= 1):             
                compressed += str(count)
            compressed += string[i+1]    # if two chars are not same, add char at next index
            count = 1
    if (count >= 1):
        compressed += str(count)
    else:
        compressed += string[i]
    return compressed

print(string_compressor('aabcccccaaa'))
```
